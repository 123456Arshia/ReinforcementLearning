# Demonstrating the Effectiveness of the ε-Greedy Strategy in Multi-Armed Bandit Problems

In this notebook, we will explore the **ε-greedy** (epsilon-greedy) strategy, a fundamental approach in reinforcement learning for balancing exploration and exploitation. We'll simulate a multi-armed bandit environment, implement agents using different ε values, and visualize their performance to understand how ε influences decision-making.

**Objectives:**
- Understand the ε-greedy strategy.
- Implement a multi-armed bandit environment.
- Compare the performance of agents with varying ε values.
- Visualize cumulative rewards, average rewards, and optimal action rates.


```python
# Install necessary libraries (uncomment if not already installed)
# !pip install numpy matplotlib

import numpy as np
import matplotlib.pyplot as plt
```

## Multi-Armed Bandit Environment

A multi-armed bandit is a simple reinforcement learning problem where an agent must choose between multiple actions (arms) to maximize its rewards. Each arm has an unknown probability of yielding a reward. The agent's goal is to identify and exploit the arm with the highest reward probability while balancing exploration of other arms.


```python
class Bandit:
    def __init__(self, k):
        self.k = k
        # True probability of reward for each arm, randomly initialized between 0 and 1
        self.probs = np.random.rand(k)

    def pull(self, arm):
        """
        Simulate pulling an arm.

        Parameters:
        - arm (int): The index of the arm to pull.

        Returns:
        - reward (int): 1 if the arm yields a reward, else 0.
        """
        return 1 if np.random.rand() < self.probs[arm] else 0
```

## ε-Greedy Agent

The ε-greedy strategy dictates that the agent explores (chooses a random arm) with probability ε and exploits (chooses the best-known arm) with probability (1 - ε). This balance helps the agent discover the optimal arm while leveraging known rewarding arms.


```python
class EpsilonGreedyAgent:
    def __init__(self, k, epsilon):
        """
        Initialize the ε-greedy agent.

        Parameters:
        - k (int): Number of arms.
        - epsilon (float): Probability of exploration (0 <= epsilon <= 1).
        """
        self.k = k
        self.epsilon = epsilon
        self.counts = np.zeros(k)  # Number of times each arm was pulled
        self.values = np.zeros(k)  # Estimated value (average reward) for each arm

    def select_arm(self):
        """
        Select an arm to pull based on the ε-greedy strategy.

        Returns:
        - arm (int): The index of the selected arm.
        """
        if self.epsilon == 0:
            # Pure exploitation: always choose the best-known arm
            return np.argmax(self.values)
        elif np.random.rand() < self.epsilon:
            # Exploration: choose a random arm
            return np.random.randint(0, self.k)
        else:
            # Exploitation: choose the best current arm
            return np.argmax(self.values)

    def update(self, arm, reward):
        """
        Update the agent's knowledge about the pulled arm.

        Parameters:
        - arm (int): The index of the pulled arm.
        - reward (int): The reward received (1 or 0).
        """
        self.counts[arm] += 1
        # Update the estimated value using the incremental formula
        n = self.counts[arm]
        value = self.values[arm]
        self.values[arm] = ((n - 1) / n) * value + (1 / n) * reward
```

## Simulation Function

The simulation function runs the interaction between the agent and the bandit over a specified number of rounds. It records the rewards obtained, the average reward per round, and the rate at which the agent selects the optimal arm.


```python
def simulate(bandit, agent, rounds):
    """
    Simulate the agent interacting with the bandit over a number of rounds.

    Parameters:
    - bandit (Bandit): The multi-armed bandit environment.
    - agent (EpsilonGreedyAgent): The agent using the ε-greedy strategy.
    - rounds (int): Number of rounds to simulate.

    Returns:
    - cumulative_rewards (np.ndarray): Cumulative rewards over time.
    - average_rewards (np.ndarray): Average reward per round over time.
    - optimal_action_rate (np.ndarray): Rate of selecting the optimal arm over time.
    """
    rewards = np.zeros(rounds)
    optimal_actions = np.zeros(rounds)

    # Determine the optimal action (arm with highest true probability)
    optimal_arm = np.argmax(bandit.probs)

    for t in range(rounds):
        arm = agent.select_arm()
        reward = bandit.pull(arm)
        agent.update(arm, reward)

        rewards[t] = reward
        if arm == optimal_arm:
            optimal_actions[t] = 1

    cumulative_rewards = np.cumsum(rewards)
    average_rewards = cumulative_rewards / (np.arange(rounds) + 1)
    optimal_action_rate = np.cumsum(optimal_actions) / (np.arange(rounds) + 1)
    return cumulative_rewards, average_rewards, optimal_action_rate
```

## Running the Experiments

We'll compare the performance of agents with different ε values:
- **ε = 0**: Pure exploitation; the agent always selects the best-known arm.
- **ε = 0.1**: Moderate exploration; the agent explores 10% of the time.
- **ε = 0.01**: Minimal exploration; the agent explores 1% of the time.

We'll visualize **Cumulative Rewards**, **Average Rewards**, and **Optimal Action Rates** to assess the effectiveness of each ε value.


```python
def main():
    np.random.seed(42)  # For reproducibility

    k = 10           # Number of arms
    rounds = 2000    # Number of rounds per simulation
    epsilons = [0, 0.1, 0.01]  # Different epsilon values to test

    bandit = Bandit(k)

    # Initialize dictionaries to store metrics for each epsilon
    cumulative_rewards_dict = {}
    average_rewards_dict = {}
    optimal_action_rate_dict = {}

    for epsilon in epsilons:
        agent = EpsilonGreedyAgent(k, epsilon)
        cumulative_rewards, average_rewards, optimal_action_rate = simulate(bandit, agent, rounds)
        cumulative_rewards_dict[epsilon] = cumulative_rewards
        average_rewards_dict[epsilon] = average_rewards
        optimal_action_rate_dict[epsilon] = optimal_action_rate

    plt.figure(figsize=(18, 5))

    # Subplot for Cumulative Rewards
    plt.subplot(1, 3, 1)
    for epsilon in epsilons:
        plt.plot(cumulative_rewards_dict[epsilon], label=f'ε = {epsilon}')
    plt.xlabel('Rounds')
    plt.ylabel('Cumulative Reward')
    plt.title('Cumulative Reward over Time')
    plt.legend()
    plt.grid(True)

    # Subplot for Average Rewards
    plt.subplot(1, 3, 2)
    for epsilon in epsilons:
        plt.plot(average_rewards_dict[epsilon], label=f'ε = {epsilon}')
    plt.xlabel('Rounds')
    plt.ylabel('Average Reward')
    plt.title('Average Reward over Time')
    plt.legend()
    plt.grid(True)

    # Subplot for Optimal Action Rate
    plt.subplot(1, 3, 3)
    for epsilon in epsilons:
        plt.plot(optimal_action_rate_dict[epsilon], label=f'ε = {epsilon}')
    plt.xlabel('Rounds')
    plt.ylabel('Optimal Action Rate')
    plt.title('Optimal Action Rate over Time')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

    # Display true probabilities and best arm
    print("\nTrue probabilities of each arm:")
    for i, p in enumerate(bandit.probs):
        print(f"Arm {i}: {p:.2f}")
    print(f"\nOptimal arm is Arm {np.argmax(bandit.probs)} with probability {np.max(bandit.probs):.2f}")

if __name__ == "__main__":
    main()
```

## Results and Analysis

After running the simulations, we obtain three plots:

1. **Cumulative Reward over Time:**
   - **ε = 0**: The agent exclusively exploits the best-known arm from the beginning. If it initially identifies the optimal arm, rewards accumulate rapidly. Otherwise, it may stick with a suboptimal arm.
   - **ε = 0.1**: The agent balances exploration and exploitation. It explores 10% of the time, allowing it to discover and exploit the optimal arm over time, leading to steady reward growth.
   - **ε = 0.01**: The agent explores minimally (1% of the time). It behaves similarly to ε = 0 but with slight exploration, potentially correcting suboptimal choices.

2. **Average Reward over Time:**
   - **ε = 0**: If the agent consistently exploits the optimal arm, the average reward stabilizes around the optimal arm's probability. If stuck with a suboptimal arm, the average reward remains lower.
   - **ε = 0.1**: The average reward gradually increases as the agent discovers and exploits the optimal arm, balancing exploration-induced fluctuations.
   - **ε = 0.01**: The average reward shows minor improvements over time compared to ε = 0, but the effect is less pronounced than with ε = 0.1.

3. **Optimal Action Rate over Time:**
   - **ε = 0**: The optimal action rate depends entirely on the initial arm selection. If the agent starts with the optimal arm, the rate remains high; otherwise, it remains low.
   - **ε = 0.1**: The optimal action rate increases over time as the agent explores and identifies the optimal arm, then exploits it more frequently.
   - **ε = 0.01**: The optimal action rate may show gradual improvement but can be similar to ε = 0 if exploration is insufficient to discover the optimal arm.

The ε-greedy strategy is a simple yet effective method for balancing exploration and exploitation in reinforcement learning. Through this simulation, we've observed:
